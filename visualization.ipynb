## Сравнение навыков (бой, инициативность, интеллект, командная работа) – радарная диаграмма для отображения средних значений.

В данном случае анализируем лучших друзей, которые весь путь прошли вместе

```python
import matplotlib.pyplot as plt
import pandas as pd
from math import pi

df = pd.read_csv("attackontitan_info_characters.csv")

# Выбираем столбцы с навыками
skill_columns = ["Навыки боя", "Инициативность", "Интеллект", "Командная работа"]

# Фильтруем по двум персонажам
characters = ["Жан Кирштейн", "Конни Спрингер"]
df_selected = df[df["name"].isin(characters)].dropna(subset=skill_columns)

# Преобразуем в числовой формат
for col in skill_columns:
    df_selected[col] = df_selected[col].apply(lambda x: int(str(x).split('/')[0]) if isinstance(x, str) else None)

# Создание фона
N = len(skill_columns)
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.set_theta_offset(pi / 2)
ax.set_theta_direction(-1)

plt.xticks(angles[:-1], skill_columns)
ax.set_rlabel_position(0)
plt.yticks([2, 4, 6, 8, 10], color="grey", size=7)
plt.ylim(0, 10)

# Добавление данных
colors = ['b', 'r']
for i, character in enumerate(characters):
    values = df_selected[df_selected["name"] == character][skill_columns].values.flatten().tolist()
    values += values[:1]
    ax.plot(angles, values, linewidth=2, linestyle='solid', label=character, color=colors[i])
    ax.fill(angles, values, colors[i], alpha=0.1)

plt.legend(loc="upper right", bbox_to_anchor=(0.1, 0.1))
plt.show()
```

## Радар чарт

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from math import pi
import ipywidgets as widgets
from IPython.display import display, clear_output

def create_radar_chart(df, character1, character2):
    fig, ax = plt.subplots(figsize=(10, 8), subplot_kw=dict(polar=True))

    skill_columns = ["Навыки боя", "Инициативность", "Интеллект", "Командная работа"]

    df_char1 = df[df["name"] == character1]
    df_char2 = df[df["name"] == character2]

    # Значения навыков для каждого персонажа
    values1 = []
    values2 = []

    for col in skill_columns:
        # Для первого персонажа
        val1 = df_char1[col].values[0]
        if isinstance(val1, str) and '/' in val1:
            values1.append(int(val1.split('/')[0]))
        elif isinstance(val1, str) and val1.strip().isdigit():
            values1.append(int(val1))
        elif isinstance(val1, (int, float)):
            values1.append(val1)
        else:
            values1.append(0)

        # Для второго персонажа
        val2 = df_char2[col].values[0]
        if isinstance(val2, str) and '/' in val2:
            values2.append(int(val2.split('/')[0]))
        elif isinstance(val2, str) and val2.strip().isdigit():
            values2.append(int(val2))
        elif isinstance(val2, (int, float)):
            values2.append(val2)
        else:
            values2.append(0)

    # Подготовка для создания графика
    N = len(skill_columns)
    angles = [n / float(N) * 2 * pi for n in range(N)]
    angles += angles[:1]

    # Добавляем первое значение в конец для замыкания графика
    values1.append(values1[0])
    values2.append(values2[0])

    ax.set_theta_offset(pi / 2)
    ax.set_theta_direction(-1)
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(skill_columns)
    ax.set_yticks([2, 4, 6, 8, 10])
    ax.set_yticklabels(['2', '4', '6', '8', '10'], color="grey")
    ax.set_ylim(0, 10)

    ax.plot(angles, values1, linewidth=2, linestyle='solid', label=character1, color='cornflowerblue')
    ax.fill(angles, values1, 'b', alpha=0.1)

    ax.plot(angles, values2, linewidth=2, linestyle='solid', label=character2, color='violet')
    ax.fill(angles, values2, 'r', alpha=0.1)

    ax.legend(loc="upper right", bbox_to_anchor=(0.1, 0.1))
    ax.set_title(f"Сравнение навыков персонажей: {character1} vs {character2}", fontsize=14)

    plt.tight_layout()
    plt.show()

# Создаем функцию для интерактивного выбора персонажей
def radar_chart_with_selection(df):
    skill_columns = ["Навыки боя", "Инициативность", "Интеллект", "Командная работа"]

    # Создаем выборку персонажей, для которых есть значения навыков
    valid_characters = []
    for index, row in df.iterrows():
        has_all_skills = True
        for col in skill_columns:
            if pd.isna(row[col]) or row[col] == '':
                has_all_skills = False
                break
        if has_all_skills:
            valid_characters.append(row['name'])

    # Создаем выпадающие списки
    character1_dropdown = widgets.Dropdown(
        options=valid_characters,
        description='Персонаж 1:',
        disabled=False,
        style={'description_width': 'initial'}
    )

    character2_dropdown = widgets.Dropdown(
        options=valid_characters,
        description='Персонаж 2:',
        disabled=False,
        style={'description_width': 'initial'}
    )

    # Задаем начальные значения, если есть хотя бы два персонажа
    if len(valid_characters) >= 2:
        character1_dropdown.value = valid_characters[0]
        character2_dropdown.value = valid_characters[1]

    # Функция обновления графика
    def update_chart(*args):
        clear_output(wait=True)
        display(character1_dropdown)
        display(character2_dropdown)
        create_radar_chart(df, character1_dropdown.value, character2_dropdown.value)

    # Привязка функции обновления к виджетам
    character1_dropdown.observe(update_chart, names='value')
    character2_dropdown.observe(update_chart, names='value')

    # Отображение начального состояния
    display(character1_dropdown)
    display(character2_dropdown)
    create_radar_chart(df, character1_dropdown.value, character2_dropdown.value)

radar_chart_with_selection(df)
```

## Цветная таблица

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("attackontitan_info_characters.csv")

def create_titan_kills_table(df):
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.axis('off')

    # Заменяем все NaN на 0 в столбце "Убито титанов В команде"
    df["Убито титанов В команде"] = df["Убито титанов В команде"].fillna(0)

    df_cleaned = df.dropna(subset=["Убито титанов В команде", "Убито титанов Соло", "Убито титанов Всего"])
    df_grouped = df_cleaned[["name", "Убито титанов В команде", "Убито титанов Соло", "Убито титанов Всего"]]
    df_grouped = df_grouped.set_index("name")

    # Получение топ-10 строк по общему количеству убитых титанов
    df_top10 = df_grouped.nlargest(10, 'Убито титанов Всего')

    # Получение данных для таблицы
    cell_text = df_top10.values
    row_labels = df_top10.index
    col_labels = df_top10.columns

    # Создание нормализованных значений для градиента
    norm_data = df_top10.values.copy().astype(float)
    for col in range(norm_data.shape[1]):
        if norm_data[:, col].max() != norm_data[:, col].min():
            norm_data[:, col] = (norm_data[:, col] - norm_data[:, col].min()) / (norm_data[:, col].max() - norm_data[:, col].min())
        else:
            norm_data[:, col] = 0.5

    copper_cmap = plt.cm.BuPu
    table = ax.table(cellText=cell_text, rowLabels=row_labels, colLabels=col_labels, loc='center', cellLoc='center')

    # Применение градиентной заливки к каждой ячейке
    for i in range(len(row_labels)):
        for j in range(len(col_labels)):
            color_val = norm_data[i, j]
            table.get_celld()[(i+1, j)].set_facecolor(copper_cmap(color_val))
            # Настройка цвета текста в зависимости от насыщенности фона
            if color_val > 0.7:
                table.get_celld()[(i+1, j)].get_text().set_color('white')

    table.auto_set_font_size(False)
    table.set_fontsize(9)
    table.scale(1.1, 1.5)
    ax.set_title("Количество убитых титанов (ТОП 10)", fontsize=14)

    plt.tight_layout()
    plt.show()

create_titan_kills_table(df)
```
